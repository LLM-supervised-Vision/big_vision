```python
import time
```

```python
from PIL import Image
from io import BytesIO

start_time = time.perf_counter()

for i in range(5):
    stem = f"{i:09d}"
    img_file = laion_tar.extractfile(stem + ".jpg")
    img = Image.open(BytesIO(img_file.read()))
    img.show()
    txt_file = laion_tar.extractfile(stem + ".txt")
    txt = txt_file.read().decode("utf-8")
    print(txt)
    json_file = laion_tar.extractfile(stem + ".json")
    json = json_file.read().decode("utf-8")
    print(json)
    img_file.close()
    txt_file.close()
    json_file.close()

end_time = time.perf_counter()

elapsed_time = end_time - start_time
print()
print(f"Elapsed time: {elapsed_time:.2f} seconds")
```

```python
from pathlib import Path

def count_files(directory):
    # Create a Path object for the directory
    path = Path(directory)
    # Use a generator expression to count files
    return sum(1 for item in path.iterdir() if item.is_file())

# Usage

number_of_files = count_files(laion_path)
print(f'Number of files: {number_of_files}')
```
