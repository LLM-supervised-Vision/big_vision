```python
import json
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
import pandas as pd
import numpy as np
from IPython.display import display, HTML

def get_final_metrics(experiment_name, experiment_path):
    metrics_path = f'{experiment_path}/big_vision_metrics.txt'

    final_metrics = {}
    with tf.io.gfile.GFile(metrics_path, 'r') as f:
        lines = f.read().split('\n')
        last_line = json.loads(lines[-2])  # Use -2 to avoid empty last line
        metrics_of_interest = [
            "z/0shot/imagenet2012_accuracy",
            "z/retr/coco_img2txt_recall@1",
            "z/retr/coco_txt2img_recall@1",
            "training_loss",
        ]
        for key in metrics_of_interest:
            if key in last_line:
                final_metrics[key] = last_line[key]
    return final_metrics

def prepare_plot_data(experiments, dataset_size):
    plot_data = []
    for exp, path in experiments.items():
        metrics = get_final_metrics(exp, path)
        model_type = 'CLIP' if 'clip' in exp.lower() else ('CLIP+GEMMA' if 'clip+gemma' in exp.lower() else 'GEMMA-Supervised')
        recap = 'Re-captioned' if 'recapt' in exp.lower() else 'Original'
        exp_type = 'DC PT' if exp.startswith('PT') else ('DC FT' if exp.startswith('FT') else 'LAION Pre-FT')
        
        for metric, value in metrics.items():
            plot_data.append({
                'Model': model_type,
                'Recap': f'{exp_type} {recap}',
                'Metric': metric,
                'Value': value,
                'Dataset': f'{dataset_size}M'
            })
    
    return pd.DataFrame(plot_data)

def plot_by_models(df, dataset_size):
    if df.empty or 'Model' not in df.columns:
        print(f"No model data to plot for dataset size {dataset_size}M")
        return

    sns.set_style("whitegrid")
    colors = sns.color_palette("husl", 5)
    
    models = df['Model'].unique()
    fig, axes = plt.subplots(1, len(models), figsize=(6 * len(models), 6))
    fig.suptitle(f'Comparison of Metrics by Model ({dataset_size}M dataset)', fontsize=16, fontweight='bold', y=1.02)

    metrics = [m for m in df['Metric'].unique() if m != 'training_loss']

    for i, model in enumerate(models):
        model_data = df[df['Model'] == model]
        
        ax = axes[i] if len(models) > 1 else axes
        x = np.arange(len(metrics))
        width = 0.2
        
        statuses = model_data['Recap'].unique()
        for j, status in enumerate(statuses):
            values = [model_data[(model_data['Metric'] == metric) & (model_data['Recap'] == status)]['Value'].values[0] 
                      if not model_data[(model_data['Metric'] == metric) & (model_data['Recap'] == status)].empty 
                      else 0 for metric in metrics]
            rects = ax.bar(x + j*width, values, width, label=status, color=colors[j], alpha=0.8)
            
            for rect in rects:
                height = rect.get_height()
                if height > 0:
                    ax.text(rect.get_x() + rect.get_width()/2., height,
                            f'{height*100:.1f}%', ha='center', va='bottom', fontsize=8)
        
        ax.set_title(model, fontsize=14, fontweight='bold', pad=10)
        ax.set_xticks(x + width * (len(statuses) - 1) / 2)
        ax.set_xticklabels(['INet', 'COCO i2t', 'COCO t2i'], rotation=0)
        ax.set_xlabel('Metrics', fontsize=10, labelpad=5)
        ax.set_ylabel('Metric Value', fontsize=10, labelpad=5)
        ax.legend(title='Status', title_fontsize='9', fontsize='7', loc='upper right', bbox_to_anchor=(1.05, 1))
        ax.set_ylim(0, 1)
        
        ax.tick_params(axis='both', which='major', labelsize=9)
        ax.grid(True, linestyle='--', alpha=0.6)
        ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))

    plt.tight_layout()
    plt.subplots_adjust(top=0.85, bottom=0.15, wspace=0.3)
    plt.show()

def plot_by_metrics(df, dataset_size):
    if df.empty or 'Metric' not in df.columns:
        print(f"No metric data to plot for dataset size {dataset_size}M")
        return

    sns.set_style("whitegrid")
    colors = sns.color_palette("husl", 3)
    
    metrics = df['Metric'].unique()
    fig, axes = plt.subplots(2, 2, figsize=(20, 16))
    fig.suptitle(f'Comparison of Models by Metric ({dataset_size}M dataset)', fontsize=16, fontweight='bold', y=0.95)

    models = df['Model'].unique()
    
    for i, metric in enumerate(metrics):
        ax = axes[i // 2, i % 2]
        
        for j, model in enumerate(models):
            model_data = df[(df['Model'] == model) & (df['Metric'] == metric)]
            
            if not model_data.empty:
                x = np.arange(len(model_data['Recap'].unique()))
                values = [model_data[model_data['Recap'] == status]['Value'].values[0] 
                          if not model_data[model_data['Recap'] == status].empty 
                          else 0 for status in model_data['Recap'].unique()]
                
                rects = ax.bar(x + j*0.25, values, 0.25, label=model, color=colors[j], alpha=0.8)
                
                for rect in rects:
                    height = rect.get_height()
                    if height > 0:
                        if metric == 'training_loss':
                            ax.text(rect.get_x() + rect.get_width()/2., height,
                                    f'{height:.3f}', ha='center', va='bottom', fontsize=7)
                        else:
                            ax.text(rect.get_x() + rect.get_width()/2., height,
                                    f'{height*100:.2f}%', ha='center', va='bottom', fontsize=7)
        
        ax.set_title(metric.split('/')[-1], fontsize=14, fontweight='bold', pad=10)
        ax.set_xticks(x + 0.25)
        ax.set_xticklabels(model_data['Recap'].unique(), rotation=45, ha='right')
        ax.set_xlabel('Status', fontsize=10, labelpad=5)
        ax.set_ylabel('Value', fontsize=10, labelpad=5)
        ax.legend(title='Model', title_fontsize='10', fontsize='8', loc='upper right')
        
        if metric != 'training_loss':
            ax.set_ylim(0, 1)
            ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))
        
        ax.tick_params(axis='both', which='major', labelsize=9)
        ax.grid(True, linestyle='--', alpha=0.6)

    plt.tight_layout()
    plt.subplots_adjust(top=0.92, hspace=0.3, wspace=0.3)
    plt.show()

def print_results_html(df):
    html_output = "<table style='border-collapse: collapse; width: 100%;'>"
    html_output += "<tr><th style='border: 1px solid black; padding: 8px;'>Model</th><th style='border: 1px solid black; padding: 8px;'>Metric</th><th style='border: 1px solid black; padding: 8px;'>Recap</th><th style='border: 1px solid black; padding: 8px;'>Value</th></tr>"
    
    for model in df['Model'].unique():
        for metric in df['Metric'].unique():
            for recap in df['Recap'].unique():
                values = df[(df['Model'] == model) & (df['Metric'] == metric) & (df['Recap'] == recap)]['Value']
                if not values.empty:
                    value = values.values[0]
                    html_output += f"<tr><td style='border: 1px solid black; padding: 8px;'>{model}</td><td style='border: 1px solid black; padding: 8px;'>{metric}</td><td style='border: 1px solid black; padding: 8px;'>{recap}</td><td style='border: 1px solid black; padding: 8px;'>{value:.4f}</td></tr>"
                else:
                    html_output += f"<tr><td style='border: 1px solid black; padding: 8px;'>{model}</td><td style='border: 1px solid black; padding: 8px;'>{metric}</td><td style='border: 1px solid black; padding: 8px;'>{recap}</td><td style='border: 1px solid black; padding: 8px;'>No data</td></tr>"
    
    html_output += "</table>"
    return html_output


def visualize_experiments(experiments, dataset_size):
    if not experiments:
        print(f"No experiments to visualize for dataset size {dataset_size}M")
        return

    df = prepare_plot_data(experiments, dataset_size)
    if df.empty:
        print(f"No data to visualize for dataset size {dataset_size}M")
        return

    plot_by_models(df, dataset_size)
    plot_by_metrics(df, dataset_size)
    display(HTML(print_results_html(df)))

# Define all experiments
all_experiments = {
    # Pre-training experiments
    'PT_clip_recapF': 'gs://us-central2-storage/tensorflow_datasets/post-training-datacomp-recap/clip_PT_lr1e-3_wd1e-4_bs16k_epoch10_warm0.03_recapFalse_50M_10-10_0458',
    'PT_clip_recapT': 'gs://us-central2-storage/tensorflow_datasets/post-training-datacomp-recap/clip_PT_lr1e-3_wd1e-4_bs16k_epoch10_warm0.03_recapTrue_50M_10-10_1226',
    'PT_gemma-supervised_recapF': 'gs://us-central2-storage/tensorflow_datasets/mllm_ckpts/paligemma/gemma2b-partial_frozen99-0.01-drop0.0-vocab256128-gap-projT_b16-beitF-qknormF_contrastive_datacomp-recapF-50M-epoch10_dpr0.0_bs16k_s3b_lr1e-3_wd1e-4_bf16_10-09_2043',
    'PT_gemma-supervised_recapT': 'gs://us-central2-storage/tensorflow_datasets/mllm_ckpts/paligemma/gemma2b-partial_frozen99-0.01-drop0.0-vocab256128-gap-projT_b16-beitF-qknormF_contrastive_datacomp-recapT-50M-epoch10_dpr0.0_bs16k_s3b_lr1e-3_wd1e-4_bf16_10-09_2039',
    
    # Pre-FT experiments
    'Pre-FT_CLIP': 'gs://us-central2-storage/tensorflow_datasets/vit-b-16_3b_pretraining/clip_bs16384_warm0.03_lr1e-3_wd1e-4_bf16_qknorm-F_b2-0.95_12lyr_07-25_1415',
    'Pre-FT_Gemma-supervised': 'gs://us-central2-storage/tensorflow_datasets/mllm_ckpts/paligemma/gemma2b-partial_frozen99-0.01-gap_b16-F_contrastive_bs16k_s3b_lr1e-3_wd1e-4_bf16_09-01_0446',

    # Fine-tuning experiments 10M
    'FT_clip_recapF': 'gs://us-central2-storage/tensorflow_datasets/post-training-datacomp-recap/clip_s3b_bs16k_lr1e-5_wd0.0_bs16k_epoch5_warm0.03_recapFalse_10-02_1912',
    'FT_clip_recapT': 'gs://us-central2-storage/tensorflow_datasets/post-training-datacomp-recap/clip_s3b_bs16k_lr1e-5_wd0.0_bs16k_epoch5_warm0.03_recapT_10-01_1442',
    'FT_gemma-supervised_recapF': 'gs://us-central2-storage/tensorflow_datasets/post-training-datacomp-recap/gemma-supervised_s3b_bs16k_lr1e-5_wd0.0_bs16k_epoch5_warm0.03_recapFalse_10-02_1727',
    'FT_gemma-supervised_recapT': 'gs://us-central2-storage/tensorflow_datasets/post-training-datacomp-recap/gemma-supervised_s3b_bs16k_lr1e-5_wd0.0_bs16k_epoch5_warm0.03_recapTrue_10-02_1437',    
    'FT_clip+gemma_recapF': 'gs://us-central2-storage/tensorflow_datasets/post-training-datacomp-recap/clip_s3b_bs16k+gemma2b_lr1e-5_wd0.0_bs16k_epoch5_warm0.03_recapFalse_10-03_0440',
    'FT_clip+gemma_recapT': 'gs://us-central2-storage/tensorflow_datasets/post-training-datacomp-recap/clip_s3b_bs16k+gemma2b_lr1e-5_wd0.0_bs16k_epoch5_warm0.03_recapTrue_10-03_0440',

    # Fine-tuning experiments 50M
    "FT_clip_50M_recapF": "gs://us-central2-storage/tensorflow_datasets/post-training-datacomp-recap/clip_s3b_bs16k_lr1e-5_wd0.0_bs16k_epoch5_warm0.03_recapFalse_50M_10-02_2028",
    "FT_clip_50M_recapT": "gs://us-central2-storage/tensorflow_datasets/post-training-datacomp-recap/clip_s3b_bs16k_lr1e-5_wd0.0_bs16k_epoch5_warm0.03_recapTrue_50M_10-02_2028",
    'FT_clip+gemma_50M_recapF': 'gs://us-central2-storage/tensorflow_datasets/post-training-datacomp-recap/clip_s3b_bs16k+gemma2b_lr1e-5_wd0.0_bs16k_epoch5_warm0.03_recapFalse_50M_10-03_0440',
    # "FT_clip+gemma_50M_recapF": "gs://us-central2-storage/tensorflow_datasets/post-training-datacomp-recap/clip_s3b_bs16k+gemma_lr2e-5_wd1e-7_bs16k_epoch20_warm0.03_50M_recapFalse_10-04_1706",
    'FT_clip+gemma_50M_recapT': 'gs://us-central2-storage/tensorflow_datasets/post-training-datacomp-recap/clip_s3b_bs16k+gemma2b_lr1e-5_wd0.0_bs16k_epoch5_warm0.03_recapTrue_50M_10-03_0440',
}

def select_experiments(experiment_types, dataset_size):
    selected = {k: v for k, v in all_experiments.items() 
                if any(k.startswith(prefix) for prefix in experiment_types) 
                and (f'{dataset_size}M' in k if 'FT' in k and dataset_size is not None else True)}
    if not selected:
        print(f"No experiments found for types {experiment_types} and dataset size {dataset_size}M")
    return selected

# Visualize Pre-training experiments
pt_experiments = select_experiments(['PT'],None)
visualize_experiments(pt_experiments, dataset_size=50)

# # Visualize all experiments
# all_exp = select_experiments(['Pre-FT','FT'])
# visualize_experiments(all_exp, dataset_size=10)

# ft_10m_experiments = select_experiments(['FT'], 10)
# visualize_experiments(ft_10m_experiments, dataset_size=10)

# ft_50m_experiments = select_experiments(['FT'], 50)
# visualize_experiments(ft_50m_experiments, dataset_size=50)

# # To visualize all experiments:
# all_exp = select_experiments(['Pre-FT', 'FT'], None)
# visualize_experiments(all_exp, dataset_size=10)

```
